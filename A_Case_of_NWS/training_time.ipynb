```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your CSV data
data = pd.read_csv('results_old.csv')

# Plot settings
plt.figure(figsize=(10, 6))
plt.title("Training Performance over Iterations", fontsize=14, fontweight='bold')

# Plot time_s
plt.plot(data['Nf'], data['time_s'], label='Time (s)', color='tab:blue', linewidth=2, marker='o')

# Plot final_loss on secondary y-axis
plt.twinx()
plt.plot(data['Nf'], data['final_loss'], label='Final Loss', color='tab:red', linewidth=2, marker='x')

# Axis labels
plt.xlabel("Number of Function Evaluations (Nf)", fontsize=12)
plt.ylabel("Final Loss (log scale)", fontsize=12, color='tab:red')
plt.yscale('log')

# Adjust layout and grid
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()

# Show plot
plt.show()
```
